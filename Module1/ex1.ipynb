## Task 1

```python
## Task1
def force(x, k=2):
    return -k*x

def energy(x, k=2):
    return 0.5*k*(x**2)

x = np.linspace(-10,10,100)

f = force(x)
u = energy(x)


plt.plot(x,f)
plt.title("Force by distance")
plt.show()

plt.plot(x,u)
plt.title("Potential Energy by Distance")
plt.show()
```

# Task 2

```python
#time steps
t = np.linspace(0,10,11)

def velo(vt, F, m, dt=1):
    return vt + (F/m)*dt

def pos(xt,vt, dt=1):
    return xt + vt*dt

F = 3
m = 1
v = [0]
x = [0]
for i in t:
    v.append(velo(v[-1], F, m))
    x.append(pos(x[-1], v[-2]))
x.pop(0)
v.pop(0)
```

```python
#plot the results
plt.plot(t,x)
plt.title("position over time")
plt.show()

plt.plot(t,v)
plt.title("velocity over time")
plt.show()
```

# Task 3: Brownian Walk

```python
# particle class
class particle:
    #initializing origin
    def __init__(self):
        self.x = 0
        self.y = 0
        self.x_hist = [0]
        self.y_hist = [0]

    # walking one step
    def walk(self, step_size):
        self.x = self.x + np.random.choice([-step_size,0,step_size])
        self.y = self.y + np.random.choice([-step_size,0,step_size])
        self.x_hist.append(self.x)
        self.y_hist.append(self.y)

#time steps
t = np.linspace(0,1000,1001)
step_size = 1
par = particle()

for i in t:
    par.walk(step_size)

print(par.x_hist, par.y_hist, sep = "\n")


plt.plot(par.x_hist, par.y_hist)
plt.show()
```
